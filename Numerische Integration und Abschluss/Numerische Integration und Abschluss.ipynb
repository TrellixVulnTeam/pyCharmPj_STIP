<b>Angenäherte Berechnung</b><br>
Findet man keine Stammfunktion, so kann man Flächeninhalte nicht mit dem Hauptsatz berechnen. Man zerschneidet dann das Flächenstück in Streifen und denkt sich den Flächeninhalt als Summe der Streifeninhalte (Prinzip von der Zerlegungsinvarianz des Flächeninhalts). Die verschiedenen Näherungsverfahren unterscheiden sich in der Form, mit der man diese Streifen näherungsweise berechnet.

<b>Trapezregel:</b><br>
Die Trapezregel beschreibt ein mathematisches Verfahren zur numerischen Annäherung des Integrals einer Funktion f(x) im Intervall [a,b] (Numerische Quadratur). Dazu ersetzt man die Fläche unter der Kurve y = f(x) im gegebenen Intervall durch ein Trapez oder mehrere gleich breite Trapeze.

<b>Simpsonsche Regel:</b><br>
Die simpsonsche Formel ist ein Verfahren der numerischen Integration, bei dem eine Näherung zum Integral einer Funktion f(x) im Intervall [a, b] berechnet wird, indem man die schwer zu integrierende Funktion f(x) durch eine exakt integrierbare Parabel P(x) annähert. Die Simpsonregel ist damit eine sogenannte abgeschlossene Newton-Cotes-Formel.
